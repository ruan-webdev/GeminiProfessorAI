```python
# Importa as bibliotecas necessárias
!pip install -q -U google-generativeai
import google.generativeai as genai
from termcolor import colored, cprint

# Define sua chave API do Google
GOOGLE_API_KEY="SUA API KEY"
genai.configure(api_key=GOOGLE_API_KEY)

# Configurações de geração de texto
generation_config = {
  "candidate_count": 1,
  "temperature": 0.5,
}

# Configurações de segurança
safety_settings = {
    "HARASSMENT": "BLOCK_NONE",
    "HATE": "BLOCK_NONE",
    "SEXUAL": "BLOCK_NONE",
    "DANGEROUS": "BLOCK_NONE"
}

# Carrega o modelo de linguagem Gemini 1.0 Pro
model = genai.GenerativeModel(model_name='gemini-1.0-pro',
                                  generation_config=generation_config,
                                  safety_settings=safety_settings,)


# Apresentação inicial
cprint("----------------------------------------", 'blue')
cprint("###  Bem-vindo ao Professor Gemini!  ###", 'red', attrs=['bold'])
cprint("----------------------------------------\n", 'blue')
print(colored("Olá, eu sou o Professor Gemini e estou aqui para te ajudar com os seus problemas escolares.\n", 'green'))
print(colored("Escolha a opção desejada para prosseguir: \n", 'yellow'))
print(colored("1. Reforço de matérias escolares", 'cyan'))
print(colored("2. Desafio diário", 'cyan'))
print(colored("3. Auxílio de deveres escolares\n", 'cyan'))


while True:
    choose = int(input(colored("Digite a opção desejada: ", 'yellow')))
    cprint("----------------------------------------\n", 'blue')

    # Reforço de matérias escolares
    if choose == 1:
        difficulty = input(colored("Digite o nome da matéria que você sente mais dificuldade: ", 'yellow'))
        theme = input(colored("Explique em poucas palavras o que você sente mais dificuldade:\n(exemplo, se for matemática, voce pode escrever que sente dificuldade em equações, se for portugues, voce pode escrever que sente dificuldade em escrita, etc...) ", 'yellow'))
        grade = int(input(colored("Qual é o seu ano escolar? (digite um número entre 1 e 12): ", 'yellow')))
        cprint("----------------------------------------\n", 'blue')
        chat = model.start_chat(history=[])
        response = chat.send_message(f"Você agora é conhecido como professor Gemini e irá sugerir dicas e links de materiais em portugues sobre a matéria de {difficulty} para o ano escolar {grade} e quero que foque em dificuldade relacionadas a {theme}")
        print(colored("\nProfessor Gemini diz:\n", 'green'), response.text, "\n")

    # Desafio diário
    elif choose == 2:
        exercise = input(colored("Digite qual matéria voce deseja receber um desafio?: ", 'yellow'))
        grade = int(input(colored("Qual é o seu ano escolar? (digite um número entre 1 e 12): ", 'yellow')))
        cprint("----------------------------------------\n", 'blue')
        chat = model.start_chat(history=[])
        response = chat.send_message(f"Você agora é conhecido como professor Gemini e irá fazer **apenas um exercicio** sobre a materia de {exercise} e o formato da pergunta tem que ser de alternativas sem informar a resposta correta. Lembrando que o ano escolar é {grade} então use somente como base esse ano escolar. Exemplo: um aluno do 1 ano é praticamente impossivel ele saber coisas relacionadas a porcentagem ou calculo de area ")
        print(colored("\nProfessor Gemini diz:\n", 'green'), response.text, "\n")
        answer = input(colored("Digite a resposta (somente a alternativa): ", 'yellow'))
        cprint("----------------------------------------\n", 'blue')
        validation = chat.send_message(f"A resposta foi {answer} se a resposta estiver correta, informe uma mensagem de parabenização pelo acerto, senão, informe a resposta correta e explique o motivo de ser aquela resposta" )
        print(colored("\nProfessor Gemini diz:\n", 'green'), validation.text, "\n")

    # Auxílio de deveres escolares
    elif choose == 3:
        materia = input(colored("Digite o enunciado da questão que voce deseja resolver: ", 'yellow'))
        cprint("----------------------------------------\n", 'blue')
        chat = model.start_chat(history=[])
        response = chat.send_message(f"Você agora é conhecido como professor Gemini, podendo resolver qualquer tipo de questão de qualquer materia e irá resolver esse problema: {materia} apresente o resultado e a explicação sobre esse problema.")
        print(colored("\nProfessor Gemini diz:\n", 'green'), response.text, "\n")

    # Verificação para continuar ou sair
    cprint("----------------------------------------\n", 'blue')
    breakpoint = input(colored("Quer continuar? (sim ou não?) ", 'yellow'))
    cprint("----------------------------------------\n", 'blue')
    if breakpoint.lower() == "sim":
        print(colored("Escolha a opção desejada para prosseguir: \n", 'yellow'))
        print(colored("1. Reforço de matérias escolares", 'cyan'))
        print(colored("2. Desafio diário", 'cyan'))
        print(colored("3. Auxílio de deveres escolares\n", 'cyan'))
    else:
        print(colored("Professor Gemini, obrigado por usar nosso assistente. Até mais!\n", 'green'))
        break
```
